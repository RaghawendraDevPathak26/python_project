```python

import matplotlib.pyplot as plt
import numpy as np
class explo:
    def expe():
        x_axis_1=np.random.randint(10,50,4)
        myexplode = [0.3, 0, 0, 0]
        
        plt.pie(x_axis_1, explode=myexplode)
        plt.show()  
explo.expe()


```

```python
import matplotlib.pyplot as plt
import numpy as np
class subplot:
    def plot():
        x_axis_2=np.random.randint(10,50,12)
        y_axis_2=np.random.randint(50,100,12)
        z_axis_2=np.random.randint(50,100,12)
        plt.figure(x_axis_2,y_axis_2,z_axis_2)
        plt.show()
subplot.plot()

```
